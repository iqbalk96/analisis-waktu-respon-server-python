```python
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
file_path = "dataset_waktu_respon_120.xlsx"  # sesuaikan path
df = pd.read_excel(file_path)

# Ambil kolom numerik (asumsi hanya 1 kolom)
data = df.select_dtypes(include='number').iloc[:, 0]

# =========================
# 1. Ukuran Pemusatan
# =========================
mean = data.mean()
median = data.median()
mode = data.mode()

print("=== Ukuran Pemusatan ===")
print(f"Mean   : {mean:.2f}")
print(f"Median : {median:.2f}")

if len(mode) == len(data):
    print("Modus  : Tidak ada (semua nilai unik)")
else:
    print(f"Modus  : {mode.tolist()}")

# =========================
# 2. Ukuran Penyebaran
# =========================
data_min = data.min()
data_max = data.max()
range_val = data_max - data_min
variance = data.var()
std_dev = data.std()

print("\n=== Ukuran Penyebaran ===")
print(f"Min       : {data_min:.2f}")
print(f"Max       : {data_max:.2f}")
print(f"Range     : {range_val:.2f}")
print(f"Varians   : {variance:.2f}")
print(f"Std Dev   : {std_dev:.2f}")

# =========================
# 3. Visualisasi
# =========================

# Histogram
plt.figure()
plt.hist(data, bins=10)
plt.title("Histogram Waktu Respon Server")
plt.xlabel("Waktu Respon (ms)")
plt.ylabel("Frekuensi")
plt.show()

# Boxplot
plt.figure()
plt.boxplot(data)
plt.title("Boxplot Waktu Respon Server")
plt.ylabel("Waktu Respon (ms)")
plt.show()

print(data.describe())
```
